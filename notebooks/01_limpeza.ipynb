```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
#lendo arquivo
df = pd.read_csv('../data/student_productivity.csv')
```

```python
df.head()
```

```python
# analisando o shape e as colunas do df
print(df.shape)
print(df.columns)
```

```python
df.info()
```

```python
#Verificando a presença de valores duplicados e valores nulos
print(f'Número de duplicatas = {df.duplicated().sum()}\n')
print(f'Soma de valores nulos: \n{df.isnull().sum()}')
```

```python
# Unindo colunas 
#Somei as horas de estudo total e removi as respectivas colunas
df['study_hours_total'] = df['study_hours'] + df['self_study_hours'] + df['online_classes_hours']
df = df.drop(['study_hours','self_study_hours','online_classes_hours'],axis=1)
df['study_hours_total']
```

```python
#criando a coluna de tempo de tela voltado para o lazer
# aqui eu somo tanto as horas gastadas com jogos quanto com as redes sociais
df['leisure_screen_time'] = df['gaming_hours'] + df['social_media_hours']
df = df.drop(['gaming_hours','social_media_hours'],axis=1)
df['leisure_screen_time'].head()
```

```python
df['internet_quality'].sample(10)
```

```python
# etapa de encoding
internet_categorias = ['Poor','Average','Good']
academic_categorias = ['High School','Undergraduate','Postgraduate']

encoder = OrdinalEncoder(categories=[internet_categorias,academic_categorias])
df[['internet_quality','academic_level']] = encoder.fit_transform(df[['internet_quality','academic_level']])
```

```python
# verificando os dados
df.head(20)
```

```python
df.columns
```

```python
# removendo outras colunas que não serão relevantes 
df = df.drop(['student_id','gender'],axis=1)
```

```python
df.to_csv('../data/student_productivity_cleaned.csv',index=False)
```
