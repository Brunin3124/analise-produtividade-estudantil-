```python
import pandas as pd
```

```python
df = pd.read_csv("../data/student_productivity_cleaned.csv")
df.head()
```

```python
#Como é o aluno médio deste dataset
df[['age','sleep_hours', 'study_hours_total','leisure_screen_time','exam_score']].median()
```

# O que realmente impacta a nota final?
- É comum pensar que uma maior quantidade de horas estudadas garante o sucesso acadêmico. Para testar essa hipótese, utilizei a **correlação de Pearson** para descobrir quais variáveis têm uma maior relação com a nota final

```python
# checando a relação de cada variável com as notas

correlacao = pd.DataFrame(df.corr()['exam_score'])
correlacao = correlacao.sort_values(by='exam_score',ascending=False)
```

```python
correlacao_positiva = correlacao.drop('exam_score')
print(f'Fatores com maior impacto positivo nas notas:\n {correlacao_positiva.head(5)}')
```

# Sobres os impactos positivos
- Fatores como produtividade, foco e saúde mental possuem um impacto considerável no sucesso acadêmico de um estudante, superando até mesmo as horas brutas estudadas.

- **Quantidade vs Qualidade**: A análise de correlação mostra que estudar com a mente cansada e sem foco rende menos do que estudar por poucas horas, mas com qualidade

```python
print(f'Fatores com maior impacto negativo nas notas:\n {correlacao.tail(5).sort_values(by="exam_score",ascending=True)}')
```

# Sobre os impactos negativos
- Notavelmente, o fator que gera o maior impacto negativo é nível de burnout dos estudantes

- Curiosamente, o tempo que o estudante gassta em telas (redes socias e jogos) tem pouco impacto na sua performance final

- Conclusão: A sobrecarga mental afeta muito mais o desempenho de um estudante do que fatores externos como o tempo de lazer

## 3. Investigando o Vilão: O que causa o Burnout?
Na análise anterior, descobrimos que o nível de esgotamento (burnout) é o principal inimigo do desempenho acadêmico. Mas o que leva um estudante ao burnout? 

Para entender isso, vamos isolar a variável `burnout_level` e procurar suas maiores correlações. O que tem correlação **positiva** aumenta o esgotamento. O que tem correlação **negativa** age como um "aliado" contra o esgotamento mental.

```python
correlacao_burnout = pd.DataFrame(df.corr()['burnout_level']).drop('burnout_level')
correlacao_burnout = correlacao_burnout.sort_values(by='burnout_level', ascending=False)

print("Fatores que mais AUMENTAM o Burnout:")
print(correlacao_burnout.head(5))
```

```python
print("Fatores que mais REDUZEM o Burnout:")
print(correlacao_burnout.tail(5).sort_values(by='burnout_level', ascending=True))
```

> ## Insights da Análise de Burnout:
> 
> * **Impactos negativos:** A pressão de **Prazos Próximos** (0.51) é, de longe, o maior causador de estresse, seguido por tentar conciliar os estudos com um **Trabalho de Meio Período**(0.37). Outro fator interessante é que o alto tempo de **exposição geral a telas**(0.34) e o alto consumo de **Cafeína** (0.19) também contribuem para o esgotamento.
> 
> * **O que ajudam a reduzir o burnout:** O principal remédio contra o burnout é o **Sono** (-0.49). Alunos que dormem mais têm níveis de esgotamento drasticamente menores. Fazer **Exercícios Físicos** (-0.15) também atua como um fator de proteção importante.
> 
> **Conclusão de Negócio:** Se uma instituição de ensino, por exemplo, quiser melhorar as notas gerais de seus alunos, a melhor estratégia não é forçar mais horas de estudo bruto, mas sim organizar melhor o calendário de provas (evitando sobreposição de prazos) e incentivar a higiene do sono e práticas esportivas.
